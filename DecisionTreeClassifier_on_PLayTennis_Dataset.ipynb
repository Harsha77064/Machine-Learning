```python
import pandas as pd
```

```python
data=pd.read_csv('PlayTennis.csv')
```

```python
data.head()
```

```python
Outlook1={'Sunny':1,'Overcast':2,'Rain':0}
Temperature1={'Hot':1,'Mild':2,'Cool':0}
Humidity1={'High':1,'Normal':2}
Wind1={'Weak':1,'Strong':2}
PlayTennis1={'No':1,'Yes':2}
```

```python
data['Outlook'].replace(Outlook1,inplace=True)
data['Temperature'].replace(Temperature1,inplace=True)
data['Humidity'].replace(Humidity1,inplace=True)
data['Wind'].replace(Wind1,inplace=True)
data['PlayTennis'].replace(PlayTennis1,inplace=True)



```

```python
data.head()
```

```python
data.describe()
```

```python
x=data.drop("PlayTennis",axis=1)
```

```python
y=data.PlayTennis
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model)
```
